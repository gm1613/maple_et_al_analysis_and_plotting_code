## Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# import matplotlib
from matplotlib import rc
import seaborn as sns
from matplotlib.ticker import MultipleLocator

# plt.rcdefaults()
rc('font',**{'family':'sans-serif','sans-serif':['Arial']})
```

## Figure 1A: Plotting spliced and unspliced *FLC* changes 

### Input FLC Total RNA data

```python

#%% Input data

full_table = pd.read_csv('TotalRNA_hongchun_20140618_rel_UBC.txt', sep="\t",skiprows=1)

full_table = full_table.rename(columns={full_table.columns[0]: "Condition"})

nv_table = full_table.loc[0:2]

w2t0_table = full_table.loc[3:5]

w2t7_table = full_table.loc[6:8]

w4t0_table = full_table.loc[9:11]

w4t7_table = full_table.loc[12:14]

w6t0_table = full_table.loc[15:17]

w6t7_table = full_table.loc[18:20]

w8t0_table = full_table.loc[21:23]

w8t7_table = full_table.loc[24:26]
```

### Calculate levels relative to mean level at non-vernalised (NV) timepoint

```python
#%% Calclulating levels relative to NV mean

nv_mean = nv_table.drop(columns="Condition").mean(axis=0).to_frame().T

nv_table = nv_table.drop(columns="Condition").div(nv_mean.loc[0],axis='columns')

w2t0_table = w2t0_table.drop(columns="Condition").div(nv_mean.loc[0],axis='columns')

w2t7_table = w2t7_table.drop(columns="Condition").div(nv_mean.loc[0],axis='columns')

w4t0_table = w4t0_table.drop(columns="Condition").div(nv_mean.loc[0],axis='columns')

w4t7_table = w4t7_table.drop(columns="Condition").div(nv_mean.loc[0],axis='columns')

w6t0_table = w6t0_table.drop(columns="Condition").div(nv_mean.loc[0],axis='columns')

w6t7_table = w6t7_table.drop(columns="Condition").div(nv_mean.loc[0],axis='columns')

w8t0_table = w8t0_table.drop(columns="Condition").div(nv_mean.loc[0],axis='columns')

w8t7_table = w8t7_table.drop(columns="Condition").div(nv_mean.loc[0],axis='columns')

#%% Computing Mean spliced and unspliced FLC

flc_primer_list = ['FLC spliced', 'FLC unspliced']

mean_table = nv_table.mean(axis=0).to_frame().T
nv_means = mean_table.filter(flc_primer_list,axis=1)


mean_table = w2t0_table.mean(axis=0).to_frame().T
w2t0_means = mean_table.filter(flc_primer_list,axis=1)
mean_table = w2t7_table.mean(axis=0).to_frame().T
w2t7_means = mean_table.filter(flc_primer_list,axis=1)

mean_table = w4t0_table.mean(axis=0).to_frame().T
w4t0_means = mean_table.filter(flc_primer_list,axis=1)
mean_table = w4t7_table.mean(axis=0).to_frame().T
w4t7_means = mean_table.filter(flc_primer_list,axis=1)

mean_table = w6t0_table.mean(axis=0).to_frame().T
w6t0_means = mean_table.filter(flc_primer_list,axis=1)
mean_table = w6t7_table.mean(axis=0).to_frame().T
w6t7_means = mean_table.filter(flc_primer_list,axis=1)

mean_table = w8t0_table.mean(axis=0).to_frame().T
w8t0_means = mean_table.filter(flc_primer_list,axis=1)
mean_table = w8t7_table.mean(axis=0).to_frame().T
w8t7_means = mean_table.filter(flc_primer_list,axis=1)

cold_means = pd.concat([nv_means,w2t0_means,w4t0_means,w6t0_means,w8t0_means],ignore_index=True)

post_cold_means = pd.concat([nv_means,w2t7_means,w4t7_means,w6t7_means,w8t7_means],ignore_index=True)


cold_spliced_means = cold_means["FLC spliced"].to_numpy()
cold_unspliced_means = cold_means["FLC unspliced"].to_numpy()

post_cold_spliced_means = post_cold_means["FLC spliced"].to_numpy()
post_cold_unspliced_means = post_cold_means["FLC unspliced"].to_numpy()

#%% Computing Std dev tables for spliced and unspliced FLC
flc_primer_list = ['FLC spliced', 'FLC unspliced']

std_table = nv_table.std(axis=0).to_frame().T
nv_stds = std_table.filter(flc_primer_list,axis=1)

std_table = w2t0_table.std(axis=0).to_frame().T
w2t0_stds = std_table.filter(flc_primer_list,axis=1)
std_table = w2t7_table.std(axis=0).to_frame().T
w2t7_stds = std_table.filter(flc_primer_list,axis=1)

std_table = w4t0_table.std(axis=0).to_frame().T
w4t0_stds = std_table.filter(flc_primer_list,axis=1)
std_table = w4t7_table.std(axis=0).to_frame().T
w4t7_stds = std_table.filter(flc_primer_list,axis=1)

std_table = w6t0_table.std(axis=0).to_frame().T
w6t0_stds = std_table.filter(flc_primer_list,axis=1)
std_table = w6t7_table.std(axis=0).to_frame().T
w6t7_stds = std_table.filter(flc_primer_list,axis=1)

std_table = w8t0_table.std(axis=0).to_frame().T
w8t0_stds = std_table.filter(flc_primer_list,axis=1)
std_table = w8t7_table.std(axis=0).to_frame().T
w8t7_stds = std_table.filter(flc_primer_list,axis=1)

cold_stds = pd.concat([nv_stds,w2t0_stds,w4t0_stds,w6t0_stds,w8t0_stds],ignore_index=True)

post_cold_stds = pd.concat([nv_stds,w2t7_stds,w4t7_stds,w6t7_stds,w8t7_stds],ignore_index=True)


cold_spliced_stds = cold_stds["FLC spliced"].to_numpy()/np.sqrt(3)
cold_unspliced_stds = cold_stds["FLC unspliced"].to_numpy()/np.sqrt(3)

post_cold_spliced_stds = post_cold_stds["FLC spliced"].to_numpy()/np.sqrt(3)
post_cold_unspliced_stds = post_cold_stds["FLC unspliced"].to_numpy()/np.sqrt(3)
```

### Make Plots

```python
#%% Spliced and unspliced FLC plot

xv = ['NV','2W', '4W', '6W', '8W']
fig, axes = plt.subplots()

axes.errorbar(xv,cold_spliced_means,yerr=cold_spliced_stds,fmt='ks-',label='FLC spliced',markersize=6)
axes.errorbar(xv,cold_unspliced_means,yerr=cold_unspliced_stds,fmt='bs--',label='FLC unspliced',markersize=6)
plt.yscale('log')
plt.legend(frameon=False,fontsize=14)
plt.ylabel('Fold change to NV',fontsize=16)
plt.xlabel('Time',fontsize=16)
axes.set_ylim([2e-2,2])
axes.set_xlim([-1,5])
#Change fontsize of tick labels
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
# plt.legend()

#set aspect ratio to 1
ratio = 1
x_left, x_right = axes.get_xlim()
y_low, y_high = axes.get_ylim()
axes.set_aspect(abs((x_right-x_left)/(y_low-y_high))*ratio)

plt.savefig('FLC_spliced_unspliced_foldchange_cold.pdf',bbox_inches='tight')
plt.show()



#%% Spliced and unspliced FLC plot postcold

xv = ['NV','2WT7', '4WT7', '6WT7', '8WT7']
fig, axes = plt.subplots()

axes.errorbar(xv,post_cold_spliced_means,yerr=post_cold_spliced_stds,fmt='ks-',label='FLC spliced',markersize=6)
axes.errorbar(xv,post_cold_unspliced_means,yerr=post_cold_unspliced_stds,fmt='bs--',label='FLC unspliced',markersize=6)
plt.yscale('log')
plt.legend(frameon=False,fontsize=14)
plt.ylabel('Fold change to NV',fontsize=16)
plt.xlabel('Time',fontsize=16)
# axes.plot(x,totalrna_predfc,'--r',label='Fold change Total RNA /NV')
axes.set_ylim([2e-2,2])
axes.set_xlim([-1,5])
#Change fontsize of tick labels
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
# plt.legend()

#set aspect ratio to 1
ratio = 1
x_left, x_right = axes.get_xlim()
y_low, y_high = axes.get_ylim()
axes.set_aspect(abs((x_right-x_left)/(y_low-y_high))*ratio)
plt.savefig('FLC_spliced_unspliced_foldchange_postcold.pdf',bbox_inches='tight')
plt.show()

# Compute ratios of spliced to unspliced FLC
cold_ratios = cold_spliced_means/cold_unspliced_means
post_cold_ratios = post_cold_spliced_means/post_cold_unspliced_means
cold_ratios_std = cold_ratios*np.sqrt((cold_spliced_stds/cold_spliced_means)**2 + (cold_unspliced_stds/cold_unspliced_means)**2)
post_cold_ratios_std = post_cold_ratios*np.sqrt((post_cold_spliced_stds/post_cold_spliced_means)**2 + (post_cold_unspliced_stds/post_cold_unspliced_means)**2)


#%% Print ratios with errors
print('Cold ratios')
print(cold_ratios)
print('Cold ratios std')
print(cold_ratios_std)

print('Post cold ratios')
print(post_cold_ratios)
print('Post cold ratios std')
print(post_cold_ratios_std)

```

## Figure 1B: Plotting spliced and unspliced *FLC* changes in *COOLAIR* defective mutants (data from Nielsen et al., 2024)

```python
#%% Comparing spliced/unspliced changes between WT and COOLAIR mutants (Full cold time-course)
# Spliced

full_table = pd.read_csv('coolair_mutants_spliced.txt', sep="\t")

full_table = full_table.rename(columns={full_table.columns[0]: "Genotype"})

spliced_ColFRI_table = full_table.loc[0:2]

spliced_TEX1_table = full_table.loc[3:5]

spliced_TEX2_table = full_table.loc[6:8]

spliced_dCOOLAIR_table = full_table.loc[9:11]


#%% Calclulating levels relative to NV mean (Spliced)

nv_mean = spliced_ColFRI_table['NV'].mean()

spliced_ColFRI_table = spliced_ColFRI_table.drop(columns="Genotype").div(nv_mean)

nv_mean = spliced_TEX1_table['NV'].mean()

spliced_TEX1_table = spliced_TEX1_table.drop(columns="Genotype").div(nv_mean)

nv_mean = spliced_TEX2_table['NV'].mean()

spliced_TEX2_table = spliced_TEX2_table.drop(columns="Genotype").div(nv_mean)

nv_mean = spliced_dCOOLAIR_table['NV'].mean()

spliced_dCOOLAIR_table = spliced_dCOOLAIR_table.drop(columns="Genotype").div(nv_mean)

#%% Comparing spliced/unspliced changes between WT and COOLAIR mutants (Full cold time-course)
# Unspliced

full_table = pd.read_csv('coolair_mutants_unspliced.txt', sep="\t")

full_table = full_table.rename(columns={full_table.columns[0]: "Genotype"})

unspliced_ColFRI_table = full_table.loc[0:2]

unspliced_TEX1_table = full_table.loc[3:5]

unspliced_TEX2_table = full_table.loc[6:8]

unspliced_dCOOLAIR_table = full_table.loc[9:11]


#%% Calclulating levels relative to NV mean (Unspliced)

nv_mean = unspliced_ColFRI_table['NV'].mean()

unspliced_ColFRI_table = unspliced_ColFRI_table.drop(columns="Genotype").div(nv_mean)

nv_mean = unspliced_TEX1_table['NV'].mean()

unspliced_TEX1_table = unspliced_TEX1_table.drop(columns="Genotype").div(nv_mean)

nv_mean = unspliced_TEX2_table['NV'].mean()

unspliced_TEX2_table = unspliced_TEX2_table.drop(columns="Genotype").div(nv_mean)

nv_mean = spliced_dCOOLAIR_table['NV'].mean()

unspliced_dCOOLAIR_table = unspliced_dCOOLAIR_table.drop(columns="Genotype").div(nv_mean)

# Keep only first five columns (NV, 1W, 2W, 4W, 6W)
spliced_ColFRI_table = spliced_ColFRI_table.iloc[:,0:5]
spliced_TEX1_table = spliced_TEX1_table.iloc[:,0:5]
spliced_TEX2_table = spliced_TEX2_table.iloc[:,0:5]
spliced_dCOOLAIR_table = spliced_dCOOLAIR_table.iloc[:,0:5]

unspliced_ColFRI_table = unspliced_ColFRI_table.iloc[:,0:5]
unspliced_TEX1_table = unspliced_TEX1_table.iloc[:,0:5]
unspliced_TEX2_table = unspliced_TEX2_table.iloc[:,0:5]
unspliced_dCOOLAIR_table = unspliced_dCOOLAIR_table.iloc[:,0:5]


#%% Computing Mean spliced and unspliced FLC for COOLAIR mutants
spliced_ColFRI_means = spliced_ColFRI_table.mean(axis=0).to_frame().T
spliced_TEX1_means = spliced_TEX1_table.mean(axis=0).to_frame().T
spliced_TEX2_means = spliced_TEX2_table.mean(axis=0).to_frame().T
spliced_dCOOLAIR_means = spliced_dCOOLAIR_table.mean(axis=0).to_frame().T

unspliced_ColFRI_means = unspliced_ColFRI_table.mean(axis=0).to_frame().T
unspliced_TEX1_means = unspliced_TEX1_table.mean(axis=0).to_frame().T
unspliced_TEX2_means = unspliced_TEX2_table.mean(axis=0).to_frame().T
unspliced_dCOOLAIR_means = unspliced_dCOOLAIR_table.mean(axis=0).to_frame().T

#%% Computing Std error for spliced and unspliced FLC for COOLAIR mutants
spliced_ColFRI_stds = spliced_ColFRI_table.std(axis=0).to_frame().T/np.sqrt(3)
spliced_TEX1_stds = spliced_TEX1_table.std(axis=0).to_frame().T/np.sqrt(3)
spliced_TEX2_stds = spliced_TEX2_table.std(axis=0).to_frame().T/np.sqrt(3)
spliced_dCOOLAIR_stds = spliced_dCOOLAIR_table.std(axis=0).to_frame().T/np.sqrt(3)

unspliced_ColFRI_stds = unspliced_ColFRI_table.std(axis=0).to_frame().T/np.sqrt(3)
unspliced_TEX1_stds = unspliced_TEX1_table.std(axis=0).to_frame().T/np.sqrt(3)
unspliced_TEX2_stds = unspliced_TEX2_table.std(axis=0).to_frame().T/np.sqrt(3)
unspliced_dCOOLAIR_stds = unspliced_dCOOLAIR_table.std(axis=0).to_frame().T/np.sqrt(3)

```

```python
#%% Spliced and unspliced FLC plot for COOLAIR mutants

xv = ['NV','1W','2W', '4W', '6W']

# For ColFRI
fig, axes = plt.subplots()
axes.errorbar(xv,spliced_ColFRI_means.to_numpy().flatten(),yerr=spliced_ColFRI_stds.to_numpy().flatten(),fmt='ks-',label='FLC spliced',markersize=6)
axes.errorbar(xv,unspliced_ColFRI_means.to_numpy().flatten(),yerr=unspliced_ColFRI_stds.to_numpy().flatten(),fmt='bs--',label='FLC unspliced',markersize=6)
plt.yscale('log')
plt.legend(frameon=False,fontsize=14)
plt.ylabel('Fold change to NV',fontsize=16)
plt.xlabel('Time',fontsize=16)
axes.set_ylim([2e-2,2])
axes.set_xlim([-1,5])
#Change fontsize of tick labels
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.title('ColFRI',fontsize=16)
# plt.legend()

#set aspect ratio to 1
ratio = 1
x_left, x_right = axes.get_xlim()
y_low, y_high = axes.get_ylim()
axes.set_aspect(abs((x_right-x_left)/(y_low-y_high))*ratio)

plt.savefig('FLC_spliced_unspliced_foldchange_ColFRI.pdf',bbox_inches='tight')
plt.show()

# For TEX1
fig, axes = plt.subplots()
axes.errorbar(xv,spliced_TEX1_means.to_numpy().flatten(),yerr=spliced_TEX1_stds.to_numpy().flatten(),fmt='ks-',label='FLC spliced',markersize=6)
axes.errorbar(xv,unspliced_TEX1_means.to_numpy().flatten(),yerr=unspliced_TEX1_stds.to_numpy().flatten(),fmt='bs--',label='FLC unspliced',markersize=6)
plt.yscale('log')
plt.legend(frameon=False,fontsize=14)
plt.ylabel('Fold change to NV',fontsize=16)
plt.xlabel('Time',fontsize=16)
axes.set_ylim([2e-2,2])
axes.set_xlim([-1,5])
#Change fontsize of tick labels
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.title('TEX1',fontsize=16)
# plt.legend()

#set aspect ratio to 1
ratio = 1
x_left, x_right = axes.get_xlim()
y_low, y_high = axes.get_ylim()
axes.set_aspect(abs((x_right-x_left)/(y_low-y_high))*ratio)

plt.savefig('FLC_spliced_unspliced_foldchange_TEX1.pdf',bbox_inches='tight')
plt.show()

# For TEX2
fig, axes = plt.subplots()
axes.errorbar(xv,spliced_TEX2_means.to_numpy().flatten(),yerr=spliced_TEX2_stds.to_numpy().flatten(),fmt='ks-',label='FLC spliced',markersize=6)
axes.errorbar(xv,unspliced_TEX2_means.to_numpy().flatten(),yerr=unspliced_TEX2_stds.to_numpy().flatten(),fmt='bs--',label='FLC unspliced',markersize=6)
plt.yscale('log')
plt.legend(frameon=False,fontsize=14)
plt.ylabel('Fold change to NV',fontsize=16)
plt.xlabel('Time',fontsize=16)
axes.set_ylim([2e-2,2])
axes.set_xlim([-1,5])
#Change fontsize of tick labels
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.title('TEX2',fontsize=16)
# plt.legend()

#set aspect ratio to 1
ratio = 1
x_left, x_right = axes.get_xlim()
y_low, y_high = axes.get_ylim()
axes.set_aspect(abs((x_right-x_left)/(y_low-y_high))*ratio)

plt.savefig('FLC_spliced_unspliced_foldchange_TEX2.pdf',bbox_inches='tight')
plt.show()

# For dCOOLAIR
fig, axes = plt.subplots()
axes.errorbar(xv,spliced_dCOOLAIR_means.to_numpy().flatten(),yerr=spliced_dCOOLAIR_stds.to_numpy().flatten(),fmt='ks-',label='FLC spliced',markersize=6)
axes.errorbar(xv,unspliced_dCOOLAIR_means.to_numpy().flatten(),yerr=unspliced_dCOOLAIR_stds.to_numpy().flatten(),fmt='bs--',label='FLC unspliced',markersize=6)
plt.yscale('log')
plt.legend(frameon=False,fontsize=14)
plt.ylabel('Fold change to NV',fontsize=16)
plt.xlabel('Time',fontsize=16)
axes.set_ylim([2e-2,2])
axes.set_xlim([-1,5])
#Change fontsize of tick labels
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.title('dCOOLAIR',fontsize=16)
# plt.legend()

#set aspect ratio to 1
ratio = 1
x_left, x_right = axes.get_xlim()
y_low, y_high = axes.get_ylim()
axes.set_aspect(abs((x_right-x_left)/(y_low-y_high))*ratio)

plt.savefig('FLC_spliced_unspliced_foldchange_dCOOLAIR.pdf',bbox_inches='tight')
plt.show()


```

## Figure 1C: Plotting *FLC* spliced to unspliced ratios (data from Nielsen et al., 2024)

### Calculate and plot *FLC* spliced to unspliced ratios for WT and COOLAIR defective mutants at 1W and 2W

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

#%% Differences in spliced/unspliced changes between WT and COOLAIR mutants (1W, 2W)

w1_ratio_means = np.array([2.635419521, 1.393358856, 1.85665507, 1.795971949])
w1_ratio_stderr = np.array([0.205965489, 0.229830888, 0.373251831, 0.15320482])

w2_ratio_means = np.array([3.005212037,1.562409311,1.412495229,1.747205894])
w2_ratio_stderr = np.array([0.38889174,0.206022133,0.079207071,0.143250833])

#%%   

ylim_lower = 0
ylim_upper = 3.5   
xv = np.array([1, 2, 3, 4])
fig, axes = plt.subplots()

genotypes = ["ColFRI", "TEX1.0", "TEX2.0", 'dCOOLAIR']
color_list = ['black', 'gray', 'gray', 'gray']
for i in range(len(xv)):
    axes.errorbar(xv[i], w1_ratio_means[i], yerr=w1_ratio_stderr[i], fmt='s', markersize=8, c=color_list[i], label=genotypes[i])

axes.set_ylim([ylim_lower, ylim_upper])
plt.xticks(xv, genotypes, rotation=45, horizontalalignment='right')
axes.set_xlim([0.2, 4.8])
axes.yaxis.set_major_locator(MultipleLocator(1))
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
# Increase line width of box
plt.setp(axes.spines.values(), linewidth=2)
# Increase line width and length of xticks
axes.xaxis.set_tick_params(width=2, length=8)

#set aspect ratio to 1
ratio = 1 / 1.616
x_left, x_right = axes.get_xlim()
y_low, y_high = axes.get_ylim()
axes.set_aspect(abs((x_right - x_left) / (y_low - y_high)) * ratio)

plt.ylabel('Spliced to unspliced ratio', fontsize=16)
plt.title('1W', fontsize=16)
plt.savefig('spliced_unspliced_ratio_1W.pdf', bbox_inches='tight')
# Show plot
plt.show()

#%%  Repeat for 2W

ylim_lower = 0
ylim_upper = 4.2   
xv = np.array([1, 2, 3, 4])
fig, axes = plt.subplots()

genotypes = ["ColFRI", "TEX1.0", "TEX2.0", 'dCOOLAIR']
color_list = ['black', 'gray', 'gray', 'gray']
for i in range(len(xv)):
    axes.errorbar(xv[i], w2_ratio_means[i], yerr=w2_ratio_stderr[i], fmt='s', markersize=8, c=color_list[i], label=genotypes[i])

axes.set_ylim([ylim_lower, ylim_upper])
plt.xticks(xv, genotypes, rotation=45, horizontalalignment='right')
axes.set_xlim([0.2, 4.8])
axes.yaxis.set_major_locator(MultipleLocator(1))
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
# Increase line width and length of xticks
axes.xaxis.set_tick_params(width=2, length=8)
# Increase line width of box
plt.setp(axes.spines.values(), linewidth=2)


#set aspect ratio to 1
ratio = 1 / 1.616
x_left, x_right = axes.get_xlim()
y_low, y_high = axes.get_ylim()
axes.set_aspect(abs((x_right - x_left) / (y_low - y_high)) * ratio)

plt.ylabel('Spliced to unspliced ratio', fontsize=16)
plt.title('2W', fontsize=16)
plt.savefig('spliced_unspliced_ratio_2W.pdf', bbox_inches='tight')
# Show plot
plt.show()
```

## Figure 1C: Perform statistical comparisons between Col*FRI* and *COOLAIR* defective mutants

## Figure S12: *FLC* spliced and unspliced changes in the early cold phase (NV to 7 days)

```python
#%% Timecourse data for FLC spliced and unspliced: NV to 7d


full_table = pd.read_csv('rbloomer_flc_timecourse_NV-7D_relUBC_NVrepscombined.txt', sep="\t")

full_table = full_table.rename(columns={full_table.columns[0]: "Condition"})

spliced_table = full_table[full_table.Primer == 'FLC-spl'].drop(columns="Primer")

unspliced_table = full_table[full_table.Primer == 'FLC-uns'].drop(columns=["Condition","Primer"]).reset_index(drop=True)

full_table = pd.concat([spliced_table,unspliced_table],axis=1,ignore_index=True)

full_table.columns = ["Condition","Spliced","Unspliced"]
nv_table = full_table.loc[0:3]

h1_table = full_table.loc[4:6]

h2_table = full_table.loc[7:9]

h4_table = full_table.loc[10:12]

h6_table = full_table.loc[13:15]

h12_table = full_table.loc[16:18]

d1_table = full_table.loc[19:21]

d2_table = full_table.loc[22:24]

d4_table = full_table.loc[25:27]

d7_table = full_table.loc[28:30]

#%% Calclulating levels relative to NV mean

nv_mean = nv_table.drop(columns="Condition").mean(axis=0).to_frame().T



def rel_nv(input_table):
    input_table = input_table.drop(columns="Condition").div(nv_mean.loc[0],axis='columns')
    return input_table

nv_table = rel_nv(nv_table)

h1_table = rel_nv(h1_table)

h2_table = rel_nv(h2_table)

h4_table = rel_nv(h4_table)

h6_table = rel_nv(h6_table)

h12_table = rel_nv(h12_table)

d1_table = rel_nv(d1_table)

d2_table = rel_nv(d2_table)

d4_table = rel_nv(d4_table)

d7_table = rel_nv(d7_table)

#%%
def mean_table(in_table):
    out_table = in_table.mean(axis=0).to_frame().T
    return out_table

nv_mean = mean_table(nv_table)

h1_mean = mean_table(h1_table)

h2_mean = mean_table(h2_table)

h4_mean = mean_table(h4_table)

h6_mean = mean_table(h6_table)

h12_mean = mean_table(h12_table)

d1_mean = mean_table(d1_table)

d2_mean = mean_table(d2_table)

d4_mean = mean_table(d4_table)

d7_mean = mean_table(d7_table)

#%%
def std_table(in_table):
    out_table = in_table.std(axis=0).to_frame().T
    return out_table

nv_std = std_table(nv_table)
    
h1_std = std_table(h1_table)

h2_std = std_table(h2_table)

h4_std = std_table(h4_table)

h6_std = std_table(h6_table)

h12_std = std_table(h12_table)

d1_std = std_table(d1_table)

d2_std = std_table(d2_table)

d4_std = std_table(d4_table)

d7_std = std_table(d7_table)

#%%

data_means = pd.concat([nv_mean,h1_mean,h2_mean,h4_mean,h6_mean,h12_mean,d1_mean,d2_mean,d4_mean,d7_mean],ignore_index=True)
data_stds = pd.concat([nv_std,h1_std,h2_std,h4_std,h6_std,h12_std,d1_std,d2_std,d4_std,d7_std],ignore_index=True)

data_spliced_means = data_means["Spliced"].to_numpy()
data_unspliced_means = data_means["Unspliced"].to_numpy()

data_spliced_stds = data_stds["Spliced"].to_numpy()/np.sqrt(3)
data_unspliced_stds = data_stds["Unspliced"].to_numpy()/np.sqrt(3)
```
